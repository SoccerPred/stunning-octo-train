```python
import pandas as pd

fbref_2010 = pd.read_csv('fbref/2010_fbref.csv',header=1)
fbref_2014 = pd.read_csv('fbref/2014_fbref.csv',header=1)
fbref_2018 = pd.read_csv('fbref/2018_fbref.csv',header=1)

display(fbref_2010.head(5),fbref_2014.head(5),fbref_2018.head(5))
```

fbref_2018 has more columns that the others

```python
fbref_2018 = fbref_2018.iloc[:,:-9]
fbref_2018.head(5)
```

```python
# concatenate them
df_concat = pd.concat([fbref_2010,fbref_2014, fbref_2018])
df_concat
```

```python
fbref_2014.rename(columns={' Pl':'Pl'},inplace=True)
fbref_2018.rename(columns={' Pl':'Pl'},inplace=True)
```

```python
df_concat = pd.concat([fbref_2010,fbref_2014, fbref_2018])
df_concat
```

```python
df_all = df_concat.groupby('Squad').mean().reset_index()
df_all
```

```python
df_all.info()
```

```python

df_all['Poss'].fillna(df_all['Poss'].mean(),inplace=True)
```

```python
df_all.mean().values
```

```python
average_row = df_all.mean().values
```

```python
df_all.loc['mean'] = df_all.mean()
df_all
```
