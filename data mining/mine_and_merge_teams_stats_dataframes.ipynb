```python
import pandas as pd

df2014 = pd.read_html('https://fbref.com/en/comps/1/1709/stats/2014-FIFA-World-Cup-Stats')
df2010 = pd.read_html('https://fbref.com/en/comps/1/19/stats/2010-FIFA-World-Cup-Stats')
df2018 = pd.read_html('https://fbref.com/en/comps/1/stats/FIFA-World-Cup-Stats')
```

### Processing df2010_teams_stats

```python
df2010_teams = df2010[0].copy()
df2010_teams.head()
```

```python
df2010_teams.columns = df2010_teams.columns.droplevel(0)
df2010_teams.rename(columns={'# Pl' : 'Pl'},inplace=True)
df2010_teams['Squad'] = df2010_teams['Squad'].str[3:]
df2010_teams.replace(' England','England',inplace=True)
df2010_teams.head()
```

### Processing df2014_teams_stats

```python
df2014_teams = df2014[0].copy()
df2014_teams.head()
```

```python
df2014_teams.columns = df2014_teams.columns.droplevel(0)
df2014_teams.rename(columns={'# Pl' : 'Pl'},inplace=True)
df2014_teams['Squad'] = df2014_teams['Squad'].str[3:]
df2014_teams.replace(' England','England',inplace=True)
df2014_teams.head()
```

### Processing df2018_teams_stats

```python
df2018_teams = df2018[0].copy()
df2018_teams.head()
```

```python
df2018_teams.columns = df2018_teams.columns.droplevel(0)
df2018_teams.rename(columns={'# Pl' : 'Pl'},inplace=True)
df2018_teams['Squad'] = df2018_teams['Squad'].str[3:]
df2018_teams.replace(' England','England',inplace=True)
df2018_teams.head()
```

df2018_teams has more columns that the others

```python
df2018_teams = df2018_teams.iloc[:,:-9]
df2018_teams.head(5)
```

```python
# concatenate them
df_concat = pd.concat([df2010_teams,df2014_teams, df2018_teams])
df_concat
```

```python
df_concat.columns
```

```python
df_concat.columns = ['Squad', 'Pl', 'Age', 'Poss', 'MP', 'Starts', 'Min', '90s', 'Gls',
       'Ast', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'Gls.1', 'Ast.1', 'G+A',
       'G-PK.1', 'G+A-PK']
df_concat.head()
```

```python
df_all = df_concat.groupby('Squad').mean().reset_index()
df_all
```

```python
df_all.info()
```

```python
df_all = df_all.replace('IR Iran','Iran')
df_all
```

```python
df_all.info()
```

```python

df_all['Poss'].fillna(df_all['Poss'].mean(),inplace=True)
```

```python
df_all.mean().values
```

```python
average_row = df_all.mean().values
```

```python
df_all.loc['mean'] = df_all.mean()
df_all
```

```python

df_all.rename({df_all.index[-1]:50}, inplace=True)
df_all
```

```python
df_all.to_csv('team_avg_statistics.csv')
```

<b>

MP : matches played

PKatt : penalty kicks attempted
