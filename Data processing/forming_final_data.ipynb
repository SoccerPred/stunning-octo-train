```python
#df_rank_2022.drop(['Unnamed: 0','Squad size','Confederation'],axis=1,inplace=True)

df_rank_2022.drop(['Squad size','Confederation'],axis=1,inplace=True)
```

```python
df_rank_2022
```

```python
df_rank_home = df_rank_2022.rename(columns={'Nation':'home_team.name','Avg. age':'home_team_Avg. age','Total value':'home_team.Total value','Points':'home_team.Points','Rank':'home_team.Rank'}
                    )
df_rank_away = df_rank_2022.rename(columns={'Nation':'away_team.name','Avg. age':'away_team. age','Total value':'away_team.Total value','Points':'away_team.Points','Rank':'away_team.Rank'}
                    )
```

```python
df_rank_away
```

```python

new_df = pd.merge(df_cleaned_stats,df_rank_home, on='home_team.name')
new_df
```

```python
new_df = pd.merge(new_df,df_rank_away, on='away_team.name')
new_df
```

```python
new_df.to_csv('final_df.csv')
```
