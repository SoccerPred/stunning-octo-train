```python
#importing necessary libraries
import pandas as pd
import numpy as np

#Display all columns for data inspecting
pd.set_option('display.max_columns', 500)
```

```python
#read the team average statistics collected from https://fbref.com/ 
team_stats = pd.read_csv('team_avg_statistics_including_friendly.csv')

team_stats = team_stats.drop('Unnamed: 0',axis=1)
team_stats.head()
```

```python
#https://www.kaggle.com/datasets/martj42/international-football-results-from-1872-to-2017
#read the international matches results from 2010 to 2018
hist_matches =  pd.read_csv('results.csv')
hist_matches = hist_matches.drop('Unnamed: 0',axis=1)
hist_matches
```

```python
#keep only important columns
hist_matches = hist_matches[['date','home_team','away_team','tournament','home_score','away_score','winning_team','goal_difference']]
```

```python
#check for team names to make sure team names are the same in all datasets to avoid error during merging
hist_matches['home_team'].value_counts().index.sort_values()[200:]
```

```python
#replace Korea Republic to South Korea
hist_matches.replace({'Korea Republic':'South Korea'},inplace=True)
```

```python
#check for team names in the statistics dataset
team_stats['Squad'].value_counts().index.sort_values()[100:]
```

```python
#replacing different team names 
team_stats.replace({'United States':'USA'},inplace=True)
team_stats.replace({'Korea Republic':'South Korea'},inplace=True)
```

### Add result column for W for home_team win, L for away_team win and D for Draw

```python
# create a list of our conditions
conditions = [
    (hist_matches['winning_team'] == hist_matches['home_team']),
    (hist_matches['winning_team'] == hist_matches['away_team'])
    ]

# create a list of the values we want to assign for each condition
values = ['W', 'L']

# create a new column and use np.select to assign values to it using our lists as arguments
hist_matches['result'] = np.select(conditions, values,'D')
hist_matches
```

```python
#read the fifa rank dataset from https://www.transfermarkt.com/statistik/weltrangliste
fifa_rank = pd.read_csv('Data processing/Fifa_Rank.csv')
fifa_rank = fifa_rank.drop('Unnamed: 0',axis=1)
fifa_rank
```

```python
#check for team names
fifa_rank['Nation'].value_counts().index.sort_values()[150:]
```

```python
#replace the only different team name
fifa_rank.replace({'U. A. E.':'United Arab Emirates'},inplace=True)
```

```python
#adding the statistics data to both teams and make home and away datasets to merge later
df_rank_home = fifa_rank.rename(columns={'Nation':'home_team','Avg. age':'home_team_Avg. age','Total value':'home_team.Total value','Points':'home_team.Points','Rank':'home_team.Rank'}
                    )
df_rank_away = fifa_rank.rename(columns={'Nation':'away_team','Avg. age':'away_team. age','Total value':'away_team.Total value','Points':'away_team.Points','Rank':'away_team.Rank'}
                    )
```

#### Merge the teams rank data with the match results in one dataframe

```python

new_df = pd.merge(hist_matches,df_rank_home, on='home_team')
new_df
```

```python
new_df = pd.merge(new_df,df_rank_away, on='away_team')
new_df
```

```python
#Drop unnecessary columns
new_df.drop(['Confederation_x','Confederation_y','Squad size_y','Squad size_x'],axis=1,inplace=True)
```

```python
new_df
```

#### Add the teams stats data to the home and away teams and merge them with data we formed in the previous step

```python
#adding prefix for away team 
df_team_away = team_stats.add_prefix('away_')
df_team_away.rename(columns={'away_Squad':'away_team'},inplace=True)
df_team_away.head(1)
```

```python
#adding prefix for home team 
df_team_home = team_stats.add_prefix('home_')
df_team_home.rename(columns={'home_Squad':'home_team'},inplace=True)
df_team_home.head(1)
```

```python
df_all = pd.merge(new_df,df_team_home, on='home_team')
df_all = pd.merge(df_all , df_team_away , on='away_team')
df_all
```

```python
#Drop the Winning team column and keep the result column which indicates if home time Wins, Lose or Draw with (W,L,D)
df_all.drop('winning_team',axis=1,inplace=True)
df_all['date'] = pd.to_datetime(df_all['date'])
df_all.info()
```

```python
#set match result column for 1 when home team win , 2 when home team lose, and 0 if draw, as the model will understand only integers
df_all['match_result'] = df_all['result'].apply(lambda x: 1 if x == 'W' else 2 if x== 'L' else 0)
df_all.head(5)
```

```python
#drop the result column and keep the new match result
df_all.drop('result',axis=1,inplace=True)
```

#### Sort the columns so hom team data come first followerd by away team data

```python
df_all.columns
```

```python
sorter = ['date','tournament','home_team','home_score','home_team.Rank','home_team_Avg. age','home_team.Total value', 'home_team.Points',
         'home_Pl', 'home_Age', 'home_Poss', 'home_MP', 'home_Starts','home_Min', 'home_90s', 'home_Gls', 'home_Ast', 'home_G-PK', 'home_PK',
       'home_PKatt', 'home_CrdY', 'home_CrdR', 'home_Gls.1', 'home_Ast.1','home_G+A', 'home_G-PK.1', 'home_G+A-PK',
         'away_team','away_score','away_team.Rank','away_team. age', 'away_team.Total value', 'away_team.Points', 'away_Pl', 'away_Age',
       'away_Poss', 'away_MP', 'away_Starts', 'away_Min', 'away_90s','away_Gls', 'away_Ast', 'away_G-PK', 'away_PK', 'away_PKatt',
       'away_CrdY', 'away_CrdR', 'away_Gls.1', 'away_Ast.1', 'away_G+A','away_G-PK.1', 'away_G+A-PK','goal_difference', 'match_result'
         ]
```

```python
df_all = df_all[sorter]
df_all.columns
```

```python
df_all.info()
```

```python
#impute the possition null values with average value
df_all['home_Poss'].fillna(df_all['home_Poss'].mean(),inplace=True)
df_all['away_Poss'].fillna(df_all['away_Poss'].mean(),inplace=True)
#df_all.drop(['home_Poss','away_Poss'],axis=1,inplace=True)
```
